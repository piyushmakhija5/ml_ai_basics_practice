```python
df_smoker = df_cardio[df_cardio.smoke==1]
print(df_smoker.shape)
df_non_smoker = df_cardio[df_cardio.smoke==0]
print(df_non_smoker.shape)
```

```python
## blood_pressure
df_smoker_bp_hi = df_smoker.ap_hi
df_smoker_bp_lo = df_smoker.ap_hi
df_non_smoker_bp_hi = df_non_smoker.ap_hi
df_non_smoker_bp_lo = df_non_smoker.ap_hi


## cholesterol
df_smoker_chol = df_smoker.cholesterol
df_non_smoker_chol = df_non_smoker.cholesterol


## weight
df_smoker_weight = df_smoker.weight
df_non_smoker_weight = df_non_smoker.weight
```

```python
## cholesterol
### Smoker
mean_smoker_chol = df_smoker_chol.mean()
std_smoker_chol = df_smoker_chol.std()
se_smoker_chol = std_smoker_chol/math.sqrt(len(df_smoker_chol))
smoker_chol_t_star = sp.stats.t.ppf(0.975, len(df_smoke_chol)-1)
print("Smoker", mean_smoker_chol, std_smoker_chol, se_smoker_chol, smoker_chol_t_star)

## non-smoker
mean_non_smoker_chol = df_non_smoker_chol.mean()
std_non_smoker_chol = df_non_smoker_chol.std()
se_non_smoker_chol = std_non_smoker_chol/math.sqrt(len(df_non_smoker_chol))
non_smoker_chol_t_star = sp.stats.t.ppf(0.975, len(df_non_smoke_chol)-1)
print("Non Smoker", mean_non_smoker_chol, std_non_smoker_chol, se_non_smoker_chol, non_smoker_chol_t_star)

## Final check
smoker_ci_plus = mean_smoker_chol + smoker_chol_t_star*se_smoker_chol
smoker_ci_minus = mean_smoker_chol - smoker_chol_t_star*se_smoker_chol
print(smoker_ci_minus,smoker_ci_plus)
non_smoker_ci_plus = mean_non_smoker_chol + non_smoker_chol_t_star*se_non_smoker_chol
non_smoker_ci_minus = mean_non_smoker_chol - non_smoker_chol_t_star*se_non_smoker_chol
print(non_smoker_ci_minus,non_smoker_ci_plus)
```

```python
## weight
### Smoker
mean_smoker = df_smoker_weight.mean()
std_smoker = df_smoker_weight.std()
se_smoker = std_smoker/math.sqrt(len(df_smoker_weight))
smoker_t_star = sp.stats.t.ppf(0.975, len(df_smoker_weight)-1)
print("Smoker", mean_smoker, std_smoker, se_smoker, smoker_t_star)

## non-smoker
mean_non_smoker = df_non_smoker_weight.mean()
std_non_smoker = df_non_smoker_weight.std()
se_non_smoker = std_non_smoker/math.sqrt(len(df_non_smoker_weight))
non_smoker_t_star = sp.stats.t.ppf(0.975, len(df_non_smoker_weight)-1)
print("Non Smoker", mean_non_smoker, std_non_smoker, se_non_smoker, non_smoker_t_star)

## Final check
smoker_ci_plus = mean_smoker + smoker_t_star*se_smoker
smoker_ci_minus = mean_smoker - smoker_t_star*se_smoker
print(smoker_ci_minus,smoker_ci_plus)
non_smoker_ci_plus = mean_non_smoker + non_smoker_t_star*se_non_smoker
non_smoker_ci_minus = mean_non_smoker - non_smoker_t_star*se_non_smoker
print(non_smoker_ci_minus,non_smoker_ci_plus)
```

```python
df_italy = df_covid[df_covid.location=='Italy']
df_italy = df_italy[df_italy.date>='2020-02-28'][df_italy.date<'2020-03-20']
df_italy['cum_cases'] = df_italy.new_cases.cumsum()
df_italy['days_passed'] = df_italy.index - df_italy.index[0]
df_italy
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def func(x, a, b):
    return a * np.exp(b * x)

x = df_italy.days_passed
y = df_italy.cum_cases
yn = y + 0.2*np.random.normal(size=len(x))

popt, pcov = curve_fit(func, x, yn)
```

```python
plt.figure()
plt.plot(x, yn, 'ko', label="Original Noised Data")
plt.plot(x, func(x, *popt), 'r-', label="Fitted Curve")
plt.legend()
plt.show()
```
